```python
# Import pustaka yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# 1. Memuat Dataset Iris
iris = load_iris()
X, y = iris.data, iris.target  # Fitur dan label
print(f"Fitur: {iris.feature_names}")
print(f"Label: {iris.target_names}")
```

```python
# 2. Mengeksplorasi Data
print(f"Jumlah data: {len(X)}")
print(f"Contoh data fitur pertama: {X[0]}")
print(f"Label pertama: {iris.target_names[y[0]]}")
```

```python
# 3. Membagi Dataset Menjadi Data Latih dan Data Uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)


print(f"Jumlah data latih: {len(X_train)}")
print(f"Jumlah data uji: {len(X_test)}")
```

```python
# 4. Membuat Model Machine Learning (K-Nearest Neighbors)
knn = KNeighborsClassifier(n_neighbors=3)

# 5. Melatih Model
knn.fit(X_train, y_train)

# 6. Mengevaluasi Model
# Memprediksi label untuk data uji
y_pred = knn.predict(X_test)

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAkurasi model: {accuracy * 100:.2f}%")
```

```python
# 7. Analisis Performa Model
print("\nLaporan Klasifikasi:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

print("Matriks Kebingungan:")
print(confusion_matrix(y_test, y_pred))
```

```python
# 8. Visualisasi Scatter Plot (Menggunakan Sepal Length dan Sepal Width)
plt.figure(figsize=(8, 6))
for i, label in enumerate(iris.target_names):
    plt.scatter(X[y == i, 0], X[y == i, 1], label=label)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot Dataset Iris (Sepal Features)')
plt.legend()
plt.show()
```

```python
# 9. Prediksi Data Baru
sample_data = [[5.1, 3.5, 1.4, 0.2], [6.7, 3.1, 4.7, 1.5]]  # Dua data sampel
predictions = knn.predict(sample_data)

for i, sample in enumerate(sample_data):
    print(f"Data sampel {sample}: Prediksi = {iris.target_names[predictions[i]]}")
```
